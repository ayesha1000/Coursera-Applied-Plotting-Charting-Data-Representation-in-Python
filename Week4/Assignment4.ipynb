```python
import pandas as pd
import numpy as np


df2010 = pd.read_csv('2010.csv')
df2011=pd.read_csv('2011.csv')
df2012=pd.read_csv('2012.csv')
df2013=pd.read_csv('2013.csv')
df2014=pd.read_csv('2014.csv')
df2015=pd.read_csv('2015.csv')
df2016=pd.read_csv('2016.csv')
df2017=pd.read_csv('2017.csv')
df2018=pd.read_csv('2018.csv')
df2019=pd.read_csv('2019.csv')

#Pakistan record
dfpak=pd.Series(name='Pakistan')
dfpak['2010']=df2010[df2010['Winner']=='Pakistan'].shape[0]
dfpak['2011']=df2011[df2011['Winner']=='Pakistan'].shape[0]
dfpak['2012']=df2012[df2012['Winner']=='Pakistan'].shape[0]
dfpak['2013']=df2013[df2013['Winner']=='Pakistan'].shape[0]
dfpak['2014']=df2014[df2014['Winner']=='Pakistan'].shape[0]
dfpak['2015']=df2015[df2015['Winner']=='Pakistan'].shape[0]
dfpak['2016']=df2016[df2016['Winner']=='Pakistan'].shape[0]
dfpak['2017']=df2017[df2017['Winner']=='Pakistan'].shape[0]
dfpak['2018']=df2018[df2018['Winner']=='Pakistan'].shape[0]
dfpak['2019']=df2019[df2019['Winner']=='Pakistan'].shape[0]
dfpak

#NewZealand record
dfzea=pd.Series(name='New Zealand')
dfzea['2010']=df2010[df2010['Winner']=='New Zealand'].shape[0]
dfzea['2011']=df2011[df2011['Winner']=='New Zealand'].shape[0]
dfzea['2012']=df2012[df2012['Winner']=='New Zealand'].shape[0]
dfzea['2013']=df2013[df2013['Winner']=='New Zealand'].shape[0]
dfzea['2014']=df2014[df2014['Winner']=='New Zealand'].shape[0]
dfzea['2015']=df2015[df2015['Winner']=='New Zealand'].shape[0]
dfzea['2016']=df2016[df2016['Winner']=='New Zealand'].shape[0]
dfzea['2017']=df2017[df2017['Winner']=='New Zealand'].shape[0]
dfzea['2018']=df2018[df2018['Winner']=='New Zealand'].shape[0]
dfzea['2019']=df2019[df2019['Winner']=='New Zealand'].shape[0]
dfzea

#England record
dfeng=pd.Series(name='England')
dfeng['2010']=df2010[df2010['Winner']=='England'].shape[0]
dfeng['2011']=df2011[df2011['Winner']=='England'].shape[0]
dfeng['2012']=df2012[df2012['Winner']=='England'].shape[0]
dfeng['2013']=df2013[df2013['Winner']=='England'].shape[0]
dfeng['2014']=df2014[df2014['Winner']=='England'].shape[0]
dfeng['2015']=df2015[df2015['Winner']=='England'].shape[0]
dfeng['2016']=df2016[df2016['Winner']=='England'].shape[0]
dfeng['2017']=df2017[df2017['Winner']=='England'].shape[0]
dfeng['2018']=df2018[df2018['Winner']=='England'].shape[0]
dfeng['2019']=df2019[df2019['Winner']=='England'].shape[0]
df = pd.concat([dfpak,dfzea,dfeng], axis=1)
#df1=df.transpose()
df.reset_index()
df
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns
fig=plt.figure()

x_labels = df.index

y_labels=df['Pakistan']
y_labels1=df['England']
y_labels2=df['New Zealand']


sns.pointplot(x = x_labels, y = y_labels,  data = df,label='Pakistan',linewidth = '',color='blue',scale=0.7)
sns.pointplot(x = x_labels, y = y_labels1,  data = df, label='England',linewidth = '',color='green',scale=0.7)
sns.pointplot(x = x_labels, y = y_labels2,  data = df, label='New Zealand',linewidth = '',color='orange',scale=0.7)
plt.gca().yaxis.grid(which = 'major', color = '#e1e1e1', linestyle = '--', linewidth = 0.5)
plt.title('Comparison of performance of countries in One Day \nInternationals in Cricket in last 10 years')
plt.xlabel('Year')
plt.ylabel('No of matches won')
plt.gca().spines['bottom'].set_color('#e1e1e1')
plt.gca().spines['bottom'].set_linestyle('--')

```
