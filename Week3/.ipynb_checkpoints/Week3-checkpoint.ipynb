# Histograms

```python
fig ,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,sharex=True)
axs=[ax1,ax2,ax3,ax4]
for n in range(0,len(axs)):
    samplesize=10**(n+1)
    sample=np.random.normal(loc=0.0,scale=1.0,size=samplesize)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(samplesize))
```

```python
# repeat with number of bins set to 100
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
plt.figure()
Y=np.random.normal(loc=0.0,scale=1.0,size=10000)
X=np.random.random(size=10000)
plt.scatter(X,Y)
```

```python
plt.figure()
Y=np.random.normal(loc=0.0,scale=1.0,size=10000)
X=np.random.random(size=10000)
plt.scatter(X,Y,alpha=0.25,color='r')
```

```python
#use gridspec to partition the figure into subplots
import matplotlib.gridspec as gridspec
plt.figure()
gspec=gridspec.GridSpec(3,3)
tophist=plt.subplot(gspec[0,1:])
sidehist=plt.subplot(gspec[1:,0])
bottomhist=plt.subplot(gspec[1:,1:])

```

```python
Y=np.random.normal(loc=0.0,scale=1.0,size=10000)
X=np.random.random(size=10000)
bottomhist=plt.scatter(X,Y)
tophist.hist(X,bins=100,color='orange')
s=sidehist.hist(Y,bins=100,orientation='horizontal')
```

```python
tophist.clear()
tophist.hist(X,bins=100,normed=True)
sidehist.clear()
sidehist.hist(Y,bins=100,orientation='horizontal',normed=True)
sidehist.invert_xaxis()
```

```python
#change axis limit
for ax in [tophist,sidehist]:
    ax.set_xlim(0,1)
```

# Box And Whisker Plots

```python
import pandas as pd
normal_sample=np.random.normal(loc=0.0,scale=1.0,size=10000)
random_sample=np.random.random(size=10000)
gamma_sample=np.random.gamma(2,size=10000)
df=pd.DataFrame({'normal':normal_sample,
                'random':random_sample,
                 'gamma':gamma_sample})
```

```python
df
```

```python
df.describe()
```

```python
plt.figure()
_=plt.boxplot(df['normal'],whis='range')
```

```python
plt.clf()
```

```python
_=plt.boxplot([df['normal'],df['random'],df['gamma']],whis='range')
```

```python
plt.figure()
plt.hist(df['gamma'],bins=100)
```
